```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import os
```

```python
drug = "PLX4720_451Lu"
```

```python
path = "./baselineData/SCAD_data" + '/' + drug + "/target_data/tp4k"
```

```python
n = 3
```

train

```python
all_train_X = pd.read_csv(path + '/' + 'X_train_target.csv',index_col=0)
all_train_X
```

```python
all_train_Y = pd.read_csv(path + '/' + 'Y_train_target.csv',index_col=0)
all_train_Y
```

```python
import random
```

```python
random.seed(5)
```

```python
sample_0 = random.sample(all_train_Y[all_train_Y["response"] == 0].index.tolist(), n)
sample_1 = random.sample(all_train_Y[all_train_Y["response"] == 1].index.tolist(), n)
```

```python
sample_0
```

```python
sample_1
```

```python
labeled_train_Y = all_train_Y.loc[sample_0 + sample_1]
labeled_train_Y
```

```python
labeled_train_X = all_train_X.loc[:,labeled_train_Y.index.tolist()]
labeled_train_X
```

```python
unlabeled_train_Y = all_train_Y.drop(sample_0 + sample_1,axis=0)
unlabeled_train_Y
```

```python
unlabeled_train_X = all_train_X.loc[:,unlabeled_train_Y.index.tolist()]
unlabeled_train_X
```
